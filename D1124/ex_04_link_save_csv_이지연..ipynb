### [윤동주 시 csv로 저장하기]
    - 웹사이트 주소 : "https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC?page=2&sort=title#section1"
    - CSV 파일로 저장

[1] 모듈로딩 <hr>

```python
from urllib.request import urlopen
from urllib.request import Request  
from bs4 import BeautifulSoup
import pandas as pd
```

[2] 웹에서 데이터 다운로드 <hr>

```python
WEB_URL = "https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC?page=2&sort=title#section1"

## Request 객체에 User-Agent 헤더 추가
req = Request( WEB_URL, 
        headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'})

## 다운로딩 
URL = urlopen(req)

## 로우데이터만 가져오기 => byte 타입의 데이터
URL_DATA = URL.read()
```

```python
BS=BeautifulSoup(URL_DATA, 'html.parser')
```

[3] 윤동주 시 추출 및 CSV로 저장 <hr>

```python
contents = BS.find_all('div', class_ = 'mw-content-ltr mw-parser-output')[0]

content = contents.find_all('li')

name = []
links = []

for c in content[4:]:
    cc = c.find('a')
    name.append(c.get_text())
    links.append("https://ko.wikisource.org/" + cc.get("href"))
```

```python
POEM_LIST = []

for i in links:
    WEB_URL = i

    req = Request( WEB_URL, 
            headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'})

    URL = urlopen(req)

    URL_DATA = URL.read()

    bs=BeautifulSoup(URL_DATA, 'html.parser')

    poem = bs.find('div', class_="prp-pages-output")
    text = poem.get_text(separator="\n", strip=True)

    POEM_LIST.append(text)

df = pd.DataFrame({
    '작품명': name,
    '내용': POEM_LIST
})
```

```python
df.to_csv('윤동주시.csv', index=False, encoding='utf-8-sig')
```
