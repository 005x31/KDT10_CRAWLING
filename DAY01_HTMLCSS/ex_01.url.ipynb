```python
# from urllib import request as req                 ## - request 모듈 전체
from urllib.request import urlopen, urlretrieve     ## - request 모듈 내 함수 2개만
import os                                           ## - 파일 경로 처리 모듈
```

[2] URL에 해당하는 데이터 로딩 <hr>

```python
## ----------------------------------------------------------------------------
## [2-1] urlretrieve 경우 ===> url의 데이터를 파일로 저장
## ----------------------------------------------------------------------------
IMG_URL = r'https://search.pstatic.net/common/?src=http%3A%2F%2Fblogfiles.naver.net%2FMjAyNTA5MDlfMTQz%2FMDAxNzU3NDA4NDE0NTU3.seJpzOi-Y30kWdMJIUyvGUztM4uRk2BP1ogSCxBIxogg.dHk7beCCBKPPvb_WRrrAS_vtfM6yGxcAEnioHPJIL-0g.JPEG%2F02.jpg&type=sc960_832'
SAVE_FILE = './img.jpg'

f_path, msg = urlretrieve(IMG_URL, SAVE_FILE)

if os.path.exists(SAVE_FILE):
    print(f"f_path => {f_path}" ) 
    print(f"msg => {msg}" )
else:
    print("URL 체크 바랍니다. 저장 실패 ")
```

```python
## ----------------------------------------------------------------------------
## [2-2] urlopen의 경우 ===> url 데이터를 저장
##                          http.client.HTTPResponse 객체에 담아서 변환
## ----------------------------------------------------------------------------
WEB_URL = r'https://www.naver.com/'

## imgReq = urlopen(IMG_URL)
webReq = urlopen(WEB_URL)

print(f"imgReq => {webReq}" )
print(f"속성 => {webReq.headers}" )
print(f"매서드 => {webReq.read()}")
```

```python
## 이미지 로우 데이터 읽어서 저장

## - 웹에서 가져오기
resObj = imReq = urlopen(IMG_URL)
## - 바이너리 데이터 추출
raw_data = resObj.read()
## - 저장할 이미지 파일명
SAVE_FILE = './img2.jpg'

## - 바이너리 모드로 파일 open & write
with open(SAVE_FILE, mode='wb') as f:
    f.write(raw_data)
    print("저장되었습니다.")
```

```python
## ----------------------------------------------------------------------------
## 텍스트 로우 데이터 읽어서 저장
## ----------------------------------------------------------------------------

## URL과 저장 경로 지정
DATA_URL = r'https://www.naver.com/'

## - ekdnsfhem & 데이터 읽어들이기
res = urlopen(DATA_URL)
data = res.read()

## - 바이너리 모드로 파일 open & write
text = data.decode('utf-8')
print("type(text) =", type(text))        ## 타입 체크
print(text)

```
