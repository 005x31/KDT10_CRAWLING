- BS4 활용해서 원하는 정보 추출 및 DataFrame 저장

[실습] <hr>
- 모든 과일 목록 추출  
- 모든 야채 목록 추출
- 색상별로 과일 및 야채 목록 추출
- 생산지별로 과일 및 야채 목록 추출
- 모든 과일과 야채를 DataFrame에 저장
    * 이름, 색상, 생산지, 종류 항목 저장

[1] 모듈 로딩 <hr> 

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup  
```

```python
## markup str ==> 객체화
html_str = """<html>
<body>
<div id="main-goods" role="page">
  <h1>과일과 야채</h1>
  <ul id="fr-list">
    <li class="red green" data-lo="ko">사과</li>
    <li class="purple" data-lo="us">포도</li>
    <li class="yellow" data-lo="us">레몬</li>
    <li class="yellow" data-lo="ko">오렌지</li>
  </ul>
  <ul id="ve-list">
    <li class="white green" data-lo="ko">무</li>
    <li class="red green" data-lo="us">파프리카</li>
    <li class="black" data-lo="ko">가지</li>
    <li class="black" data-lo="us">아보카도</li>
    <li class="white" data-lo="cn">연근</li>
  </ul>
</div>
</body>
</html>"""
mbs = BeautifulSoup(html_str, "html.parser")
```

```python
## 모든 과일목록 추출
fruits = mbs.find_all(id="fr-list")

for p in fruits:
    print(p.text)
```

```python
## 모든 야채 목록 추출
vegitable = mbs.find_all(id="ve-list")

for p in vegitable:
    print(p.text)
```

```python
## 색상별로 과일 및 야채 목록 추출
colors = ['red green', 'white green', 'yellow', 'purple', 'black', 'white']

for color in colors:
    color_tags = mbs.find_all(class_ = color)

    print(f"== {color} ==")
    for tag in color_tags:
        print(tag.text)
        
    print("\n")
    
```

```python
## 생산지별로 과일 및 야채 목록 추출
locations = ['ko', 'us', 'cn']

for loc in locations:
    loc_tags = mbs.find_all(attrs={'data-lo': loc})

    print(f"== {loc} ==")
    for tag in loc_tags:
        print(tag.text)

    print("\n")
```

```python
## 모든 과일과 야채를 DataFrame에 저장
##    * 이름, 색상, 생산지, 종류 항목 저장

import pandas as pd

name = []
color = []
loc = []
kind = []

all_fr_ve = mbs.find_all('li')
for i in all_fr_ve:
    name.append(i.text)
    color.append(i['class'])
    loc.append(i['data-lo'])

    if i in mbs.find_all(id="fr-list")[0].find_all('li'):
        kind.append('과일')
    else:
        kind.append('야채')

pd.DataFrame({'이름':name, '색상':color, '생산지':loc, '종류':kind})
```
